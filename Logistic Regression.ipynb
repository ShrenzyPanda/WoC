```python
IM = np.identity(26)
y_class = Train[:,0]
y_list = []
for i in range(Train.shape[0]) :
    a = y_class[i]-1
    y_list.append(IM[a,:])
Y = np.array(y_list)
print(Y,'\n',Y.shape)
```

```python
X_Test1 = Test[:,1:Test.shape[1]]
print(X_Test1,'\n',X_Test1.shape)
```

```python
X_Test = X_Test1/255
X_Test
```

```python
def logistic_regression(iterations,learning_rate,X,Y,X_Test):
    theta = np.zeros((X.shape[1],26))
    m = X.shape[0]
    for i in range(iterations):
        Z = np.dot(X,theta)
        hyp = 1/(1+np.exp(-Z))
        cost = (np.sum((Y)*np.log(hyp)+(1-Y)*np.log(1-hyp)))*(-1/m)
        grad = np.dot(X.T,(hyp-Y))
        theta = theta - (learning_rate/m) * (grad)
        plt.scatter(i,cost)
    print(cost)
    Y_P1 = np.dot(X_Test,theta)
    Y_Prob = 1/(1+np.exp(-Y_P1))
    return Y_Prob
    
```

```python
Y_Prob=logistic_regression(500,1.2,X,Y,X_Test)
```

```python
Y_Pred2 = np.argmax(Y_Prob,axis=1)
Y_Pred3 = Y_Pred2 + 1
Y_Pred3
```
