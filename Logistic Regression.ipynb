```python
IM = np.identity(26) #One-Hot Encoding / One vs All Multiclass Classifier
y_class = Train[:,0]
y_list = []
for i in range(Train.shape[0]) :
    a = y_class[i]-1
    y_list.append(IM[a,:])
Y = np.array(y_list)
pd.DataFrame(Y)
```

```python
X1 = Train[:,1:Train.shape[1]] #Extracting Training Inputs
pd.DataFrame(X1)
```

```python
X = X1/255 #Feature Scaling for Training Data
pd.DataFrame(X)
```

```python
X_Test1 = Test[:,1:Test.shape[1]] #Extracting Testing inputs
pd.DataFrame(X_Test1)
```

```python
X_Test = X_Test1/255 #Feature Scaling for Testing Data
pd.DataFrame(X_Test)
```

```python
def logistic_regression(iterations,learning_rate,X,Y,X_Test): #Logistic Regression Function
    theta = np.zeros((X.shape[1],26)) #Initializing Parameters
    m = X.shape[0]
    for i in range(iterations):
        Z = np.dot(X,theta) #Multiplying Parameters
        hyp = 1/(1+np.exp(-Z)) #Hypothesis : Sigmoid Function
        cost = (np.sum((Y)*np.log(hyp)+(1-Y)*np.log(1-hyp)))*(-1/m) #Cost Function : Binary Cross Entropy
        grad = np.dot(X.T,(hyp-Y)) #Gradient
        theta = theta - (learning_rate/m) * (grad) #Parameter Simultaneous Updation
        plt.scatter(i,cost) #Plotting Gradient Descent
    print(cost)
    Y_P1 = np.dot(X_Test,theta)
    Y_Prob = 1/(1+np.exp(-Y_P1)) #Probability
    return Y_Prob
    
```

```python
Y_Prob=logistic_regression(2000,1.2,X,Y,X_Test)
```

```python
pd.DataFrame(Y_Prob) #Respective Probabilities
```

```python
Y_Pred1 = np.argmax(Y_Prob,axis=1) #Predicting the Alphabet
Y_Pred = Y_Pred1 + 1
pd.DataFrame(Y_Pred)
```

```python
Y_True = Test[:,0]
pd.DataFrame(Y_True)
```

```python
Y_Diff=(Y_Pred-Y_True).tolist() #Taking Difference between Predicted and Actual Alphabet numbers
print(Y_Diff)
```
