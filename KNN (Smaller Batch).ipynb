```python
Train = X_Train0.to_numpy()
Test1 = X_Test0.to_numpy()
```

```python
#Python Indexing for Testing Data Set. Max Examples at a time = 100  
Start = 34
End = 134
```

```python
M = End - Start #Testing Examples
```

```python
Test = Test1[Start:End] #Extracting Testing Input
```

```python
def K_Calc(Test): #Determining Nearest Neighbours(K)
    K = int(((Test.shape[0])**(0.5))//1)
    if K % 2 == 0 : #K must be odd
        K = K+1
    return K
```

```python
K = K_Calc(Test)
K
```

```python
def dist(Test,Train): #Calculating Distances
    l1 = []
    dict1 = dict()
    for p in range(M) :
        for q in range(Train.shape[0]) :
            diff = Test[p,:] - Train[q,:]
            r = np.sqrt(np.sum(diff**2)) #One Testing Example with all Training Examples
            dict1[r] = q #Assigning the row index with distances for further reference
            if q%88000 == 0 :
                print(r,q,p)
        l1.append(dict1)
        dict1 = dict()
    return l1
```

```python
Res = dist(Test,Train)
```

```python
Res
```

```python
len(Res)
```

```python
def sort(Res): #Sorting : Ascending Order of Distance
    l = []
    for p in range(len(Res)):
        Sort = sorted(Res[p].items())
        l.append(Sort)
        Sort = []
    return l
```

```python
Res_Sort = sort(Res)
```

```python
len(Res_Sort)
```

```python
def k_neighbours(Res_Sort) : #Findinf K Nearest Neighbours
    l1 = []
    l2 = []
    for p in range(len(Res_Sort)):
        for q in range(K):
            l1.append(Res_Sort[p][q][1])
        l2.append(l1)
        l1 =[]
    return l2     
```

```python
Res_K = k_neighbours(Res_Sort)
Res_K
```

```python
print(len(Res_K),len(Res_K[1]))
```

```python
Y_Train1 = KNN_Train[0].to_numpy() #Extracting Training Outputs
Y_Train = Y_Train1[:]
print(Y_Train.shape)
pd.DataFrame(Y_Train)
```

```python
def alphabet(Res_K): #Finding Output of Training Inputs who are the Nearest Neighbours 
    l1 = []
    l2 = []
    for p in range(len(Res_K)):
        for q in range(K):
            r1 = Res_K[p][q] 
            l1.append(Y_Train[r1])
        l2.append(l1)
        l1 = []
    return l2        
```

```python
Res_Letter = alphabet(Res_K)
Res_Letter
```

```python
def prediction(Res_Letter): #Finding maximum occuring training output within a dataset and predicting it as the output for Testing Data
    Pred = []
    for p in range(len(Res_Letter)):
        r = max(Res_Letter[p],key=Res_Letter[p].count)
        Pred.append(r)
    Y_Pred = np.array(Pred)
    return Y_Pred
```

```python
Y_Pred = prediction(Res_Letter)
pd.DataFrame(Y_Pred)
```
