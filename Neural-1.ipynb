```python
IM = np.identity(26)
y_class = Train[:,0]
y_list = []
for i in range(Train.shape[0]) :
    a = y_class[i]-1
    y_list.append(IM[a,:])
Y = np.array(y_list)
print(Y,'\n',Y.shape)
```

```python
X = (Train[:,1:Train.shape[1]])/255
print(X,X.shape)
```

```python
X_Test = (Test[:,1:Test.shape[1]])/255
print(X_Test,X_Test.shape)
```

```python
def neural_network(iterations,learning_rate,X,Y,X_Test):
    m = X.shape[0]
    Weight1 = np.random.rand(X.shape[1],26)
    Bias1 = np.random.rand(X.shape[0],1)
    Weight2 = np.random.rand(26,26)
    Bias2 = np.random.rand(X.shape[0],1)
    for i in range(iterations):
        #Forward Propagation
        Z2 = np.dot(X,Weight1) + Bias1
        A2 = 1/(1+np.exp(-Z2))
        Z3 = np.dot(A2,Weight2) + Bias2
        A3 = 1/(1+np.exp(-Z3))
        #Backward Propagation
        Del3 = A3 - Y
        der = A2*(1-A2)
        Del2 = (np.dot(Del3,Weight2.T))*der
        dW2 = (np.dot(A2.T,Del3))/m
        dW1 = (np.dot(X.T,Del2))/m
        dB2 = (np.sum(Del3))/m
        dB1 = (np.sum(Del2))/m
        #Updating Hyperparameters
        Weight1 = Weight1 - learning_rate*dW1
        Weight2 = Weight2 - learning_rate*dW2
        Bias1 = Bias1 - learning_rate*dB1
        Bias2 = Bias2 - learning_rate*dB2
    Y_T1 = np.dot(X_Test,Weight1)
    Y_Prob1 = 1/(1+np.exp(-Y_T1))
    Y_T2 = np.dot(Y_Prob1,Weight2)
    Y_Prob = 1/(1+np.exp(-Y_T2))
    return Y_Prob
```

```python
Y_Prob = neural_network(500,0.5,X,Y,X_Test)
```
