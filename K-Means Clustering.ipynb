```python
def KMC(X,K,iterations) :
    centroid = X[np.random.choice(X.shape[0],K,replace=False)] #Initializing Centroid
    distance1 = np.ones((19,1))
    for i in range(iterations):
        for p in range(X.shape[0]):
            diff = (centroid - X[p,:])**2 
            dist = (np.sqrt(np.sum(diff,axis = 1))).reshape(19,1) #Calculating Distance
            distance1 = np.append(distance1,dist,axis=1)
        distance = np.delete(distance1,0,axis=1)
        clus = np.argmin(distance,axis=0) #Obtaining cluster for each dataset
        for q in range(K):
            if len(X[clus==q]) != 0 : #If a cluster has no element its centroid remains same
                centroid[q,:] = (X[clus == q]).mean(axis=0) #Finiding New Centroid
        distance1 = np.ones((19,1))
    clusters = clus + 1 #Cluster Number (Starting from 1)
    return clusters, centroid
```

```python
Y_Pred, Centroid = KMC(X,K,100)
pd.DataFrame(Y_Pred)
```

```python
def inter_nuc_dist(Centroid): #Finding Inter Nuclear Distances
    dist1 = np.ones((K,1))
    for p in range(K):
        diff = (Centroid[p,:] - Centroid)**2
        dist = np.sqrt(np.sum(diff,axis=1).reshape(K,1))
        dist1 = np.append(dist1,dist,axis=1)
    Distances = np.delete(dist1,0,axis=1)
    return Distances
```

```python
Inter_ND = inter_nuc_dist(Centroid)
```

```python
Min_Inter_ND = np.amin(Inter_ND_Sort1) #Finding overall Minimum Inter Nuclear Distance
Min_Inter_ND
```

```python
def intra_nuc_dist(X,Y_Pred,Centroid): #Calculating Intra Nuclear Distances
    l1 = []
    for p in range(K):
        if X[Y_Pred==p].shape[0] != 0 :
            diff = (X[Y_Pred==p] - Centroid[p,:])**2
            dist1 = np.sqrt(np.sum(diff,axis=1))
            dist2 = (2/X[Y_Pred==p].shape[0])*(np.sum(dist1))
            l1.append(dist2)
    return l1
```

```python
Intra_ND = intra_nuc_dist(X,Y_Pred,Centroid)
```
