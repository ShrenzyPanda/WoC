```python
def dist(X):
    centroid = np.random.rand(19,X.shape[1])
    clus_matrix1 = np.ones((19,1))
    for p in range(X.shape[0]):
        diff = (centroid - X[p,:])**2
        dist = (np.sqrt(np.sum(diff,axis = 1))).reshape(19,1)
        clus_matrix1 = np.append(clus_matrix1,dist,axis=1)
    clus_matrix = np.delete(clus_matrix1,0,axis=1)
    return clus_matrix
```

```python
distance = dist(X)
```
