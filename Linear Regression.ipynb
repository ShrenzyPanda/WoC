```python
def linear_regression(iterations, learning_rate, X, Y, X_Test): #Linear Regression Function
  theta = np.vstack(np.zeros(X.shape[1]))   #Initializing Parameters
  m = X.shape[0]
  for p in range(iterations) :
    hyp = np.dot(X,theta) #Hypothesis 
    cost = (np.sum((hyp-Y)**2))/(2*m) #Cost Function # a , b  Cost = a^2 + b^2 Luck = (a+b)^2 
    grad = np.dot(X.T,(hyp-Y))  #Gradient 
    theta = theta - (learning_rate/m)*grad #Parameter Simultaneous Updation
    plt.scatter(p,cost) #Plotting Gradient Descent
  Y_Test = np.dot(X_Test,theta)
  print(cost)

  return Y_Test
```
