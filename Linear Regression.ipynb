```python
Y1 = Train["label"] #Extracting Training Output
Y = (Y1.to_numpy()).reshape(Train.shape[0],1)
pd.DataFrame(Y)
```

```python
Y_True1 = Test["label"] #Extracting Testing Outputs
Y_True = (Y_True1.to_numpy()).reshape(Test.shape[0],1) 
pd.DataFrame(Y_True)
```

```python
X1 = Train.copy() #Standardization Of Training Inputs
del X1['label']
X1 = (X1 - X1.mean(axis=0))/(X1.std(axis=0))
X1
```

```python
X0 = np.ones((X1.shape[0],1)) #Bias Vector
```

```python
X2=X1.to_numpy() #Adding Bias to Training Inputs
X = np.append(X0,X2,axis=1)
pd.DataFrame(X)
```

```python
X_T1 = Test.copy() #Standardization of Testing Inputs
del X_T1['label']
X_T1 = (X_T1 - X_T1.mean(axis=0))/(X_T1.std(axis=0))
X_T1
```

```python
X0_T = np.ones((X_T1.shape[0],1)) #Creating Bias Vector and adding it to Testing Inputs
X_T2 = X_T1.to_numpy()
X_Test = np.append(X0_T,X_T2,axis=1)
pd.DataFrame(X_Test)
```

```python
def linear_regression(iterations, learning_rate, X, Y, X_Test): #Linear Regression Function
    theta = np.vstack(np.zeros(X.shape[1]))   #Initializing Parameters
    m = X.shape[0]
    for p in range(iterations) :
        hyp = np.dot(X,theta) #Hypothesis 
        cost = (np.sum((hyp-Y)**2))/(2*m) #Cost Function : Mean Squared Error  
        grad = np.dot(X.T,(hyp-Y))  #Gradient 
        theta = theta - (learning_rate/m)*grad #Parameter Simultaneous Updation
        plt.scatter(p,cost) #Plotting Gradient Descent
    Y_Test = np.dot(X_Test,theta) #Predction
    print(cost)
    return Y_Test
```

```python
Y_Pred = linear_regression(2000,0.1,X,Y,X_Test)
```

```python
def MSE(A,B): #Calculating Mean Squared Error
    m = A.shape[0]
    Diff = (A-B)**2
    R = (np.sum(Diff))/(2*m)
    return R
```
