```python
Scaled_Train = NN_Train.iloc[:,1:]/255
Scaled_Test = NN_Test.iloc[:,1:]/255
```

```python
X1 = Scaled_Train.to_numpy()
X_Test1 = Scaled_Test.to_numpy()
print(X1.shape,"\n",X_Test1.shape)
```

```python
IM = np.identity(26)
y_class = NN_Train.iloc[:,0]
y_list = []
for i in range(NN_Train.shape[0]) :
    a = y_class[i]-1
    y_list.append(IM[a,:])
Y1 = np.array(y_list)
print(Y1,'\n',Y1.shape)
```

```python
X = X1.T
X_Test = X_Test1.T
Y = Y1.T
print(X.shape,X_Test.shape,Y.shape)
```

```python
def neural_network(iterations,learning_rate,X,Y,X_Test):
    #Initialisation
    m = X.shape[1]
    n = X.shape[0]
    l = Y.shape[0]
    theta1 = 0.001*np.ones((n,100))
    bias1 = np.ones((100,1))
    theta2 = 0.001*np.ones((100,l))
    bias2 = np.ones((l,1))
#     dt1 = 0
#     dt2 = 0
#     db1 = 0
#     db2 = 0
    for i in range(iterations):
        #Forward Propagation
        Z2 = np.dot(theta1.T,X) + bias1
        A2 = 1/(1+np.exp(-Z2))
        Z3 = np.dot(theta2.T,A2) + bias2
        A3 = 1/(1+np.exp(-Z3))
        #Backward propagation
        Del3 = A3 - Y
        der = A2 * (1-A2)
        Del2 = np.dot(theta2,Del3) * der
        #Computing Gradients
        dt1 = (1/m) * np.dot(Del2,X.T)
        dt2 = (1/m) * np.dot(Del3,A2.T)
        db1 = (1/m) * (np.sum(Del2,axis=1)).reshape(100,1)
        db2 = (1/m) * (np.sum(Del3,axis=1)).reshape(l,1)
        #Assigning New Hyperparameters
        theta1 = theta1 - dt1.T * learning_rate
        theta2 = theta2 - dt2.T * learning_rate
        bias1 = bias1 - db1 * learning_rate
        bias2 = bias2 - db2 * learning_rate
    Y_T1 = np.dot(theta1.T,X_Test) + bias1
    Y_P1 = 1/(1+np.exp(-Y_T1))
    Y_T2 = np.dot(theta2.T,Y_P1) + bias2
    Y_Prob = 1/(1+np.exp(-Y_T2))
    return Y_Prob
```

```python
Y_Prob = neural_network(500,0.001,X,Y,X_Test)
Y_Prob
```

```python
Y_Pred2 = np.argmax(Y_Prob,axis=0)
Y_Pred3 = Y_Pred2 + 1
Y_Pred3
```
